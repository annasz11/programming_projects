```python
brackets = set()
for par in itertools.product([' ','(',')','((','))'], repeat=4):
    c = 0
    s = ''.join(par)
    for i in s:
        if i == '(': c += 1
        elif i == ')': c -= 1
        if c < 0: break
    if c == 0: brackets.add(par)
```
